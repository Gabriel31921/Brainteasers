This file is dedicated to this brainteaser:

*"You assist to the auction of a painting, you win the auction if your bid is higher than the painting value, and if you win, you have contacted a museum that will buy the painting for 1,5 times the actual value of the painting, so, how much will you bid?"*

We know that we don't lose money just by placing a bid, e.i. if we bid less than the value (V), we don't lose the bid, so we are not forced to bid high, but now we have to now the function that tells us if we make money or not for a bid X. We are going to assume the bids are in the closed interval [0, 100]

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0, 101)
V = np.arange(0, 101)
```

Our return function will be the next:<br><br>
R = B * 1,5 - X   -- *if X > B*<br>
R = 0  --  *if X < B*<br><br>
So, for a given X, the expected value of said X will be:<br><br>
E = - X + S * P<br>
S = The sum of every B < X times 1,5<br>
P = The probability of every B happening<br><br>
A pretty asy way of thinking about expected values is thinking about them being either a mean or an integral, depending on the variable you have, as this variable is an integer betweeen 0 and 100, its a mean, rather a distribution function over P.<br><br>
This will be done in the case that the values of the painting are not uniformly distributed, because if they are, this means that the painting has the same chance of having a value 0, or X, or every value in the middle (I exclude the cases in where the painting has a higher value than our bid because those cases are not important), so having this distribution we can conclude that the average price for the painting is X/2, 0 + 1 + 2... + X-1 + X every member of the posible values has the same probability, so the average value would be the sum divided by X+1, not gonna do all the math here, the sum would be X*(X+1) / 2, and then that divided by X+1 and simplified is X/2.

```python
E = []

for x in X:
    individual_return = x/2 * 1.5 - x
    E.append(individual_return)

plt.plot(X, E, label="Expected Return")
plt.xlabel("Bid (X)")
plt.ylabel("Expected Return")
plt.title("Expected Return vs. Bid in Auction")
plt.legend()
plt.grid()
```

Pretty easy, mostly logic. Now let's try another approach:
